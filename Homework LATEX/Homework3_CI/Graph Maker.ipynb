```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

```python
def funct(x):
    y = (1 / (statdev * np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2 / (2 * variance)))
    return y


mean = 0
statdev = 1
variance = statdev ** 2
# Range factor
factor = 3.4

a, b = -2.14, 2.14

x = np.linspace(a - statdev * factor, b + statdev * factor, 10000)
y = funct(x)
    
fig, ax = plt.subplots()
ax.plot(x, y, 'k')
ax.set_ylim(bottom=0)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

# Shaded Region (More Effective Method)
xfill = np.linspace(a - statdev * factor, b + statdev * factor, 10000)
ax.fill_between(xfill, y, 0, where=(x > a) & (x <= b), color='red', alpha='0.6')

ax.set_xticks((mean, a, b))
ax.set_xticklabels(('$\mu$', '$-2.14\sigma$', '$-2.14\sigma$'))

plt.axvline(x=mean, color='k', linestyle='-.', linewidth=0.5, ymax=0.95)

# ax.text(3, 0.35, '$\sigma = {}$'.format(statdev))

ax.set_title(r'Normal Distribution Plot')
ax.set_xlabel(r'$z$')

plt.savefig('problem1.pdf')

plt.show()
```

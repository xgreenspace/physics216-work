```python
import matplotlib.pyplot as plt
import numpy as np
import csv

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

# Y - Direction Graphs

## Position vs Time

```python
position = []
frame = []

with open('lab2 puck.csv') as file:
    csv_reader = csv.reader(file, delimiter=',')
    line_count = 0
    
    for row in csv_reader:
        if line_count == 0:
            line_count += 1
        elif row[5] == '' or row[0] == '':
            continue
        else:
            position.append(float(row[5]))
            frame.append(float(row[0]))
            

x = frame            
y = position


plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.title('Frames (1/30 second) vs Y Position (pixels)')
plt.ylabel('Position (pixels)')
plt.xlabel('Frames (1/30 second)')

y_y = plt.plot(x, y, color='k', linestyle=':', label='$y(t)$')
plt.legend()
plt.savefig('position_graphy.svg')
```

## Velocity vs Time

```python
velocity = []
frame = []

with open('lab2 puck.csv') as file:
    csv_reader = csv.reader(file, delimiter=',')
    line_count = 0
    
    for row in csv_reader:
        if line_count == 0:
            line_count += 1
        elif row[7] == '' or row[0] == '':
            line_count += 1
            continue
        elif line_count == 42:
            break
        else:
            line_count += 1
            velocity.append(float(row[7]))
            frame.append(float(row[0]))
            

x = frame            
y = velocity


plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.title('Frames (1/30 seconds) vs Y Velocity (pixels/second)')
plt.ylabel('Velocity (pixels/second)')
plt.xlabel('Frames (1/30 seconds)')

v_y = plt.plot(x, y, color='b', linestyle='-.', label='$v_{y}(t)$')
plt.legend()

plt.savefig('velocity_graphy.svg')
```
